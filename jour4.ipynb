```python
# LRU Cache (Medium) - Problem #146

# Python program to implement LRU Least Recently Used)
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.prev = None
        self.next = None


class LRUCache:
    def __init__(self, capacity):
        self.capacity = capacity
        self.cache = {}
        self.head = Node(-1, -1)
        self.tail = Node(-1, -1)
        self.head.next = self.tail
        self.tail.prev = self.head

    def add(self, node):
      
        # Add a node right after the head
        # (most recently used position).
        next_node = self.head.next
        self.head.next = node
        node.prev = self.head
        node.next = next_node
        next_node.prev = node

    def remove(self, node):
        # doubly linked list.
        prev_node = node.prev
        next_node = node.next
        prev_node.next = next_node
        next_node.prev = prev_node

    def get(self, key) :
        # Get the value for a given key
        if key not in self.cache:
            return -1

        node = self.cache[key]
        self.remove(node)
        self.add(node)
        return node.value

    def put(self, key, value):
        #Put a key-value pair into the cache.
        if key in self.cache:
            node = self.cache[key]
            self.remove(node)
            del self.cache[key]

        if len(self.cache) >= self.capacity:

            # Remove the least recently used item
            # (just before the tail)
            lru_node = self.tail.prev
            self.remove(lru_node)
            del self.cache[lru_node.key]

        # Add the new node
        new_node = Node(key, value)
        self.add(new_node)
        self.cache[key] = new_node


```

```python

#1. Valid Parentheses (Easy) - Problem #20

class Solution(object):
    def isValid(self, s):
        opcl = {"(" : ")","[" : "]", "{" : "}"}
        liste = []
        for c in s:
            if c in '([{':
                liste.append(c)
            elif len(liste) == 0 or c != opcl[liste.pop()]:
                return False
        return len(liste) == 0
```

```python
#2. Min Stack (Easy) - Problem #155



class MinStack:
  def __init__(self):
    self.stack = []

  def push(self, x):
    mn = x if not self.stack else min(self.stack[-1][1], x)
    self.stack.append([x, mn])

  def pop(self) :
    self.stack.pop()

  def top(self):
    return self.stack[-1][0]

  def getMin(self):
    return self.stack[-1][1]
```

```python
# 3. Evaluate Reverse Polish Notation (Medium) - Problem #150

class Solution:
  def evalRPN(self, tokens) :
    stack = []
    op = {
        '+': lambda a, b: a + b,
        '-': lambda a, b: a - b,
        '*': lambda a, b: a * b,
        '/': lambda a, b: int(a / b)}

    for token in tokens:
      if token in op:
        b = stack.pop()
        a = stack.pop()
        stack.append(int(op[token](a, b)))
      else:
        stack.append(int(token))

    return stack.pop()


tokens =["10","6","9","3","+","-11","*","/","*","17","+","5","+"]

sol = Solution()

print (sol.evalRPN(tokens))
```
